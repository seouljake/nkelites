```python
# duplicate changes - shape
m[(m["_merge"]=="both") & (m.duplicated(["PrimaryInstitution","OrgName"],keep=False))].sort_values(["PrimaryInstitution","OrgName"]).shape
```

```python
# unique changes - shape
m[(m["_merge"]=="both") & ~(m.duplicated(["PrimaryInstitution","OrgName"],keep=False))].sort_values(["PrimaryInstitution","OrgName"]).shape
```

```python
# duplicate changes
duplicates = m.loc[(m["_merge"]=="both") & (m.duplicated(["PrimaryInstitution","OrgName"],keep=False)),["PrimaryInstitution","OrgName","Edit_PI","Edit_OrgName"]].sort_values(["PrimaryInstitution","OrgName"])
```

```python
# unique changes - shape
uniques = m.loc[(m["_merge"]=="both") & ~(m.duplicated(["PrimaryInstitution","OrgName"],keep=False)),["PrimaryInstitution","OrgName","Edit_PI","Edit_OrgName"]].sort_values(["PrimaryInstitution","OrgName"])
```
