# import libraries

```python
import pandas as pd
import os
```

# define globals

```python
path_data_in = "../../data/2022 data/2022 data - 0 raw/NK elite new checklist.xlsx"
path_spreadsheets_folder = "../../data/2022 data/2022 data - 0 raw/NK elite spreadsheets/"
path_data_out = "../../data/2022 data/2022 data - 2 tables/leaders.csv"
```

# import data: NK elite new checklist

```python
dfnew = pd.read_excel(path_data_in).dropna(subset=["LeaderID"])
```

# rename column names for checklist

```python
newcolumns = ["LeaderID","Name","NameAlternative","CurrentPositionString","Sex","BirthDateString","BirthPlaceString","DeathReported","DeathString"]
dfnew.columns=newcolumns
dfnew
```

# import filenames from NK elite spreadsheets

```python
fnames = os.listdir(path_spreadsheets_folder)
```

```python
fnames
```

```python
data = [[x.split(" ")[0],x] for x in fnames]
```

```python
data
```

```python
dffolder = pd.DataFrame(data,columns=["LeaderID","filename"])
```

# merge filenames with NK elite master list

```python
dfmerge = dfnew.merge(dffolder,on="LeaderID",how="outer")
```

```python
dfmerge
```

# confirm 1-1 match between files and master checklist

```python
dfmerge[dfmerge.filename.isnull()]
```

```python
dfmerge[dfmerge.Name.isnull()]
```

# export master checklist as .csv table

```python
dfnew.to_csv(path_data_out,index=False)
```
