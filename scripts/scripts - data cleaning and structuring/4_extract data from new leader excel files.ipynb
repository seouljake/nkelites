# define functions

```python
def update_section(path, filename, leaderid, df, stopword):
        
    f = xlrd.open_workbook(path + filename)
    
    sheet = f.sheet_by_index(0)
    
    dfupdate = pd.DataFrame(columns = df.columns)
        
    for rowindex in range(0,sheet.nrows):
                
        row = sheet.row_values(rowindex)
        
        #print(row)

        # instantiate logical variables
        if(rowindex==0):
            startreadindex = 10000
            readcurrent = False
            readrange = False

        if((rowindex>=startreadindex) and readrange):
            readcurrent = True
            if(row[0]==""):
                #print(row)
                readcurrent = False
                readrange = False

        if(readcurrent):
            #print(row)
            row_to_append = [leaderid]+row[0:len(df.columns)-1]
            #print(row_to_append)
            
            series_to_append = pd.Series(row_to_append,index = df.columns)
            #print(series_to_append)
            
            dfupdate = dfupdate.append(series_to_append,ignore_index=True)
            
            #print(row_to_append)
            #print(series_to_append)
            #print(career_update)

        if(row[0]==stopword):
            startreadindex = rowindex+2
            readrange = True
    
    if(dfupdate.shape[0]==0):
        print("no ",stopword," for: ",leaderid)
        
    return dfupdate
```

# import excel spreadsheets 

```python
fnames = os.listdir(path)
```

```python
#fnames
```

```python
for filename in fnames:
    leaderid = filename.split(" ")[0]
    
    career_update = update_section(path, filename, leaderid, career,"경력사항")
    family_update = update_section(path, filename, leaderid, family,"가족사항")
    education_update = update_section(path, filename, leaderid, education,"학력사항")
    
    career = career.append(career_update,ignore_index=True).drop_duplicates()
    family = family.append(family_update,ignore_index=True).drop_duplicates()
    education = education.append(education_update,ignore_index=True).drop_duplicates()
```
